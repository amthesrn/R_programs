### Common coding questions implimented with R language in Menu driven format

1. Implement a simple grading system where the user enters a score, and the program outputs the corresponding grade using if-else.

2. Write R program to check if a character entered by the user is a vowel or a consonant.

3. Write a program to find the maximum of three numbers using nested if-else.

4. Write a function to check if a given year is a century year or not.

5. Write R program to find the roots of a quadratic equation using if-else.

6. Write R program that determines if a given string is a valid email address.

7. Write R program that calculates the factorial of a given number and checks if the result is even or odd using if-else.

8. Write a function that accepts a list of numbers and returns the sum of all positive numbers and the product of all negative numbers.

9. Write a function that takes a list of integers as input and returns the count of even and odd numbers.

```r
#1.grade systrem
grade_system=function(){
    grade=0
    score=as.integer(readline("\nEnter score(0-100):"))

    if (score>90){
        grade='A'
    }else if(score>70){
        grade='B'
    }else if(score>50){
        grade='C'
    }else if(score>30){
        grade='pass'
    }else{
        grade='fail'
    }

    cat('\nScore:',score,'Grade:',grade,"\n")
}
```

```r
#2.vowel_consonant_check
vowel_consonant_check = function() {
  char <- readline("\nEnter a character: ")

  if (tolower(char) %in% c("a", "e", "i", "o", "u")) {
    cat(char," is vowel\n")
  } else {
    cat(char," is consonant\n")
  }
}
```

```r
#3.maximum_of_three
maximum_of_three=function(){
  num1=as.integer(readline("\nEnter no1:"))
  num2=as.integer(readline("\nEnter no2:"))
  num3=as.integer(readline("\nEnter no3:"))

  max=0
  if (num1>num2 & num1>num3){
    max=num1
  }else if(num2>num1 & num2>num3){
    max=num2
  }else{
    max=num3
  }

  cat('\nMaximum of ',num1,num2,num3,' is: ',max )
}
```

```r
#4.century_year_checking
century_year=function(){
  year=as.integer(readline('\nEnter the year(integer of 4 digit):'))

  if (year %% 100 ==0){
    cat(year,' is a century year.')
  }else{
    cat(year,' is not a century year.')
  }
}
```

```r
#5.roots_of_quadratic_equation
roots_of_quadratic_equation=function(){
  print("\nEquation in the form: ax*2 + by + c")
  a=as.integer(readline("\nEnter the value of a:"))
  b=as.integer(readline("\nEnter the value of b:"))
  c=as.integer(readline("\nEnter the value of c:"))

  # Calculate the discriminant
  discriminant <- b^2 - 4 * a * c

  # Check if the discriminant is positive, negative, or zero
  if (discriminant > 0) {
    # Two real and distinct roots
    root1 = (-b + sqrt(discriminant)) / (2 * a)
    root2 = (-b - sqrt(discriminant)) / (2 * a)
    cat("\nRoots are real and distinct:\n")
    cat("\nRoot 1:", root1, "\n")
    cat("\nRoot 2:", root2, "\n")
    }else if (discriminant == 0) {
    # One real root (repeated)
    root = -b / (2 * a)
    cat("\nRoots are real and repeated:\n")
    cat("\nRoot:", root, "\n")
  } else {
    # Complex roots
    real_part = -b / (2 * a)
    imaginary_part = sqrt(-discriminant) / (2 * a)
    cat("\nRoots are complex:\n")
    cat("\nRoot 1:", real_part, "+", imaginary_part, "i\n")
    cat("\nRoot 2:", real_part, "-", imaginary_part, "i\n")
  }

}
```

```r
#6.check_for_valid_mail
check_for_valid_mail <- function() {
  cat("\nEnter an email id (format: string@string.com): ")
  mail <- readline()

  pattern <- "^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,}$"

  # Check if the email matches the pattern
  if (grepl(pattern, mail)) {
    cat("\n",mail, " is a valid email address.\n")
  } else {
    cat("\n",mail, " is an invalid email address.\n")
  }
}
```

```r
#7.factorial_even_odd_check
factorial_even_odd_check=function(){
  num=as.integer(readline('\nEnter a number:'))
  f=1
  factorial=0
  if (num==0 | num==1){
    factorial=1
  }else if (num<0){
    cat('\nfactorial of negative number is not possible!!!')
  }else{
    for(i in 2:num){
      f=f*i
    }
    factorial=f
  }

  cat(num,'!=', factorial )

  if (factorial %% 2==0){
    cat(factorial,' is an even number')
  }else{
   cat(factorial,' is an odd number')
  }
}
```

```r
#8.num_operation
num_operation=function(){

  num_str = readline("\nEnter numbers separated by commas (e.g., 1,2,3): ")
  num_list = as.integer(strsplit(num_str, ",")[[1]])

  sum_positive = 0
  product_negative = 1

  for (num in num_list) {
    if (num > 0) {
      sum_positive = sum_positive + num
    } else if (num < 0) {
      product_negative = product_negative * num
    }
  }

  cat('\nSum of positive numbers:',sum_positive)
  cat('\nProduct of negative numbers:',product_negative)
}
```

```r
#9.count_of_even_odd
count_of_even_odd=function(){

  num_str = readline("\nEnter numbers separated by commas (e.g., 1,2,3): ")
  num_list = as.integer(strsplit(num_str, ",")[[1]])

  even_count=0
  odd_count=0

  for (num in num_list){
    if (num %% 2==0){
      even_count= even_count +1
    }else{
    odd_count= odd_count +1
  }
  }

  cat('\ncount of even numbers in the list:',even_count)
  cat('\ncount of odd numbers in the list:',odd_count)
}
```

```r
#main menu driven block

menu <- function() {
  cat("\nMenu Driven Program\n")
  cat("1. Grade_system:\n")
  cat("2. Vowel_consonant_check:\n")
  cat("3. Maximum_of_three:\n")
  cat("4. Century_year:\n")
  cat("5. Roots_of_quadratic_equation:\n")
  cat("6. String_is_email:\n")
  cat("7. factorial_even_odd_check:\n")
  cat("8. Num_operation:\n")
  cat("9. Count_of_even_odd:\n")
  cat("10. Exit\n")
}

getOption = function() {
  #to clear the console
  flush.console()

  choice = as.numeric(readline("\nEnter your choice: "))
  return(choice)
}

while (TRUE) {
  menu()
  option = getOption()

  if (option == 1) {
    cat("\nGrade_system:\n")
    grade_system()

  } else if (option == 2) {
    cat("Vowel_consonant_check:\n")
    vowel_consonant_check()

  } else if (option == 3) {
    cat("Maximum_of_three:\n")
    maximum_of_three()

  } else if (option == 4) {
    cat("Century_year:\n")
    century_year()


  }else if (option == 5) {
    cat("Roots_of_quadratic_equation:\n")
    roots_of_quadratic_equation()


  } else if (option == 6) {
    cat("check_for_valid_mail:\n")
    check_for_valid_mail()


  } else if (option == 7) {
    cat("factorial_even_odd_check:\n")
    factorial_even_odd_check()


  } else if (option == 8) {
    cat("Num_operation:\n")
    num_operation()


  } else if (option == 9) {
    cat("Count_of_even_odd:\n")
    count_of_even_odd()


  } else if (option == 10) {
    break

  }
  else {
    cat("Invalid choice. Please enter a valid option.\n")
  }
}
```
